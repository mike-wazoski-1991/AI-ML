Pandas has two main data structures:
(i)Series : Think of it as a single column of data with labels(indices), a one-dimensional labelled array.
it can hold any data-type(string,float,integer)

(ii)DataFrame: It is a two-dimensional labelled data-structure, it consist of rows and coulmns, where each column is actually a series.



```python
#Series
import pandas as pd

# Creating a Series from a list
data = [10, 20, 30, 40]
series = pd.Series(data, index=['a', 'b', 'c', 'd'])

print(series)
```

```python
#DataFrame
import pandas as pd

# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [24, 30, 22],
    'City': ['New York', 'London', 'Paris']
}

df = pd.DataFrame(data)

print(df)
```
